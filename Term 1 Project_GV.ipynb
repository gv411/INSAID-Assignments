## 6. Data Processing & Data Normalization

```python
# Check count of '?' in normalized losses column
t1_dfAutomobile['normalized-losses'].value_counts()
t1_dfAutomobile['normalized-losses'].loc[t1_dfAutomobile['normalized-losses'] == "?"].count()
```

```python
# Calculating the mean of the data and replacing which have value '?' in normalized-losses variable
t1_nlMean = t1_dfAutomobile[~(t1_dfAutomobile['normalized-losses'] == '?')]['normalized-losses'].astype(str).astype(int).mean()
t1_dfAutomobile['normalized-losses'] = t1_dfAutomobile['normalized-losses'].astype(str).replace('?',t1_nlMean).astype(int)
t1_dfAutomobile['normalized-losses'].value_counts()
```

```python
# Calculating the mean of the data and replacing which have value '?' in bore variable
t1_dfAutomobile['bore'].loc[t1_dfAutomobile['bore'] == "?"].count()
t1_bMean = t1_dfAutomobile[~(t1_dfAutomobile['bore'] == '?')]['bore'].astype(str).astype(float).mean()
t1_dfAutomobile['bore'] = t1_dfAutomobile['bore'].astype(str).replace('?',t1_bMean).astype(float)
t1_dfAutomobile['bore'].value_counts()
```

```python
# Calculating the mean of the data and replacing which have value '?' in stroke variable
t1_dfAutomobile['stroke'].loc[t1_dfAutomobile['stroke'] == "?"].count()
t1_sMean = t1_dfAutomobile[~(t1_dfAutomobile['stroke'] == '?')]['stroke'].astype(str).astype(float).mean()
t1_dfAutomobile['stroke'] = t1_dfAutomobile['stroke'].astype(str).replace('?',t1_sMean).astype(float)
t1_dfAutomobile['stroke'].loc[t1_dfAutomobile['stroke'] == "?"].count()
```

```python
# Calculating the mean of the data and replacing which have value '?' in price variable
t1_dfAutomobile['price'].loc[t1_dfAutomobile['price'] == "?"].count()
t1_pMean = t1_dfAutomobile[~(t1_dfAutomobile['price'] == '?')]['price'].astype(str).astype(float).mean()
t1_dfAutomobile['price'] = t1_dfAutomobile['price'].astype(str).replace('?',t1_pMean).astype(float)
t1_dfAutomobile['price'].loc[t1_dfAutomobile['price'] == "?"].count()
```

```python
# Calculating the mean of the data and replacing which have value '?' in horsepower variable
t1_dfAutomobile['horsepower'].loc[t1_dfAutomobile['horsepower'] == "?"].count()
t1_hMean = t1_dfAutomobile[~(t1_dfAutomobile['horsepower'] == '?')]['horsepower'].astype(str).astype(float).mean()
t1_dfAutomobile['horsepower'] = t1_dfAutomobile['horsepower'].astype(str).replace('?',t1_hMean).astype(float)
t1_dfAutomobile['horsepower'].loc[t1_dfAutomobile['horsepower'] == "?"].count()
```

```python
# Calculating the mean of the data and replacing which have value '?' in peak-rpm variable
t1_dfAutomobile['peak-rpm'].loc[t1_dfAutomobile['peak-rpm'] == "?"].count()
t1_rMean = t1_dfAutomobile[~(t1_dfAutomobile['peak-rpm'] == '?')]['peak-rpm'].astype(str).astype(int).mean()
t1_dfAutomobile['peak-rpm'] = t1_dfAutomobile['peak-rpm'].astype(str).replace('?',t1_rMean).astype(int)
t1_dfAutomobile['peak-rpm'].loc[t1_dfAutomobile['peak-rpm'] == "?"].count()
```

```python
# Removing the 3 records which are having the value '?' in num-of-doors variable
t1_dfAutomobile['num-of-doors'].loc[t1_dfAutomobile['num-of-doors'] == '?']
t1_dfAutomobile = t1_dfAutomobile[t1_dfAutomobile['num-of-doors'] != '?']
t1_dfAutomobile['num-of-doors'].loc[t1_dfAutomobile['num-of-doors'] == '?']
```

```python
# Get Distinct values of Make
print("Count of distinct values of Make:", len(set(t1_dfAutomobile['make'])))
print("Distinct values of Make:", set(t1_dfAutomobile['make'])) 
```

```python
# Get missing data information in complete set
t1_total = t1_dfAutomobile.isnull().sum().sort_values(ascending=False)
t1_percent = ((t1_dfAutomobile.isnull().sum()/t1_dfAutomobile.isnull().count())*100).sort_values(ascending=False)
t1_missing_data = t1_pd.concat([t1_total, t1_percent], axis=1, keys=['Total', 'Percent'])
print(t1_missing_data)
```
